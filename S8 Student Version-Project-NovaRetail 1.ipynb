### V de Cramér

```python
def cramers_v(df, col1, col2):

    tabla = pd.crosstab(df[col1], df[col2])

    chi2, _, _, _ = chi2_contingency(tabla)

    n = tabla.sum().sum()

    v = np.sqrt(chi2 / (n * (min(tabla.shape) -1)))

    print(f"V de Cramér ({col1} vs {col2}): {v:.4f}") # Función para calcular V de Cramér
```

```python
cramers_v(df, 'abandono', 'tipo_dispositivo')
cramers_v(df, 'abandono', 'region')
cramers_v(df, 'abandono', 'miembro_premium') # Aplicar V de Cramér en variables relevantes

```

Observaciones V de Cramér

**abandono vs tipo_dispositivo**

Se obtuvo un V de Cramér = 0,007, lo que indica una asociación prácticamente nula entre el tipo de dispositivo y el abandono de los clientes.
El tipo de dispositivo aporta muy poca información para explicar el comportamiento de abandono en este conjunto de datos.

**abandono vs region**

Se obtuvo un V de Cramér = 0,015, lo que indica una asociación muy débil o prácticamente inexistente entre la región y el abandono.
La distribución del abandono es similar entre las distintas regiones analizadas.

**abandono vs miembro_premium**

Se obtuvo un V de Cramér = 0,120, lo que indica una asociación débil entre la condición de miembro premium y el abandono.
Aunque esta relación es más intensa que las observadas para tipo_dispositivo y region, su magnitud sigue siendo baja y no representa una asociación fuerte.


## Sección 5 - Interpretación de resultados para el negocio

### Hallazgo 1 — Compras mensuales e ingreso anual

**Evidencia visual:** El scatterplot muestra una tendencia positiva clara entre compras_mes e ingreso_anual, con baja dispersión y sin outliers relevantes.

**Evidencia numérica:** Correlación de Pearson: r = 0.967.

**Interpretación**
Existe una asociación lineal muy fuerte y positiva entre ambas variables. Los clientes que realizan un mayor número de compras tienden a registrar mayores valores de ingreso anual generado para la empresa.

**No podemos afirmar**
No puede concluirse que aumentar la cantidad de compras provoque un incremento del ingreso anual, ya que el análisis es observacional y únicamente identifica una asociación.

**Implicación de negocio**  
Esta relación sugiere que ambas variables pueden utilizarse para identificar clientes de mayor valor y complementar procesos de segmentación o análisis de comportamiento.


### Hallazgo 2 — Visitas mensuales e ingreso anual

**Evidencia visual:** El scatterplot muestra una tendencia positiva entre visitas_mes e ingreso_anual, aunque con una dispersión considerable entre las observaciones.

**Evidencia numérica:** Correlación de Spearman: ρ = 0,321.

**Interpretación** Se observa una asociación monotónica positiva de intensidad moderada entre las visitas mensuales y el ingreso anual generado. En general, los clientes que realizan un mayor número de visitas tienden a generar mayores ingresos para la empresa, aunque la variabilidad de los datos indica que este comportamiento no es uniforme en todos los clientes.

**No podemos afirmar** No puede concluirse que un aumento en el número de visitas provoque un incremento en el ingreso anual, ya que el análisis identifica únicamente una asociación y no una relación de causalidad.

**Implicación de negocio**
La relación observada sugiere que la frecuencia de visitas podría considerarse un indicador complementario para comprender el comportamiento de los clientes y analizar su contribución al ingreso anual, aunque su asociación es considerablemente menor que la encontrada entre compras_mes e ingreso_anual.
